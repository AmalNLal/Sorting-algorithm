```python
import matplotlib.pyplot as plt
import time
import random
%matplotlib notebook
```

```python
plt.rcParams["animation.html"]="jshtml"
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
fig.show()

x,y=[],[]
for i in range(100):
    y.append(i)
    x.append(random.randint(1,101))
t=len(x)

def merge(x,l):
    
    for i in range (0,t,l):
        for j in range(i,min(i+l,t)):
            for k in range(i,j):
                if x[k]>x[k+1]:
                    x[k],x[k+1]=x[k+1],x[k]
                    
        ax.clear()
        ax.bar(x,y,width=3)
        ax.set_xlabel("Ordering")
        ax.set_ylabel("Elaments")
        fig.canvas.draw()           

                
       
    while l<len(x):
        l=l*2
        merge(x,l)    

merge(x,2)

plt.close()



            
   
        
```
